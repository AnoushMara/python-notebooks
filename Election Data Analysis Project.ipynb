```python
#For data
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

#For visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from __future__ import division
```

```python
#requests module to import data from the web in text form
#StringIO to work with csv file taken from HuffPost

import requests
from StringIO import StringIO
```

```python
# csv Data source
url = "http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv"

#Getting data in text form
source = requests.get(url).text

#Cancels IO error with pandas
poll_data = StringIO(source)
```

```python
poll_df = pd.read_csv(poll_data)

#Overview
poll_df.info()
```

```python
poll_df.head()
```

```python
#Rship with affiliation and polls

sns.catplot('Affiliation', data = poll_df, kind = 'count')
```

```python
#Majority has no affiliation. 
#Comparatively Dem has greater affiliation than Rep

#Adding pop hue to get more insight

sns.catplot('Affiliation', hue = 'Population', data = poll_df, kind = 'count', aspect = 1)
```

```python
#Likely and Registered Voters have greater power

poll_df.head()
```

```python
#Averages for Obama, Romney and Undecided

#Getting overall average
avg = pd.DataFrame(poll_df.mean())
avg.drop('Number of Observations', axis = 0, inplace = True)

#Getting error
std = pd.DataFrame(poll_df.std())
std.drop('Number of Observations', axis = 0, inplace = True)

avg.plot(yerr = std, kind = 'bar')

#Doubt yerr? 
```

```python
#Romney and Obama polls are very close.
#Thus undecided voters would make a big difference
```

```python
#Putting Avg and Std dataframes together in poll_df

poll_avg = pd.concat([avg,std], axis = 1)

poll_avg.columns = ['Average', 'STD']
poll_avg
```

```python
#Since undecided voters make a big difference
#Checking the numbers if undecided voters are evenly split

#Quick plot: Time series analysis  between voter sentiment and End date
#Note: reverse chronological order, and multiple polls per end date

poll_df.plot(x = 'End Date', y = ['Obama', 'Romney', 'Undecided'], marker = 'o', linestyle = '')
```

```python
#Difference in voter sentiment with time

from datetime import datetime

poll_df['Difference'] = (poll_df.Obama - poll_df.Romney)/100
poll_df.head()

#Postive difference means leaning toward Obama
```

```python
# Set as_index=Flase to keep the 0,1,2,... index. Then we'll take the mean of the polls on that day.
poll_df = poll_df.groupby(['Start Date'],as_index=False).mean()

# Let's go ahead and see what this looks like
poll_df.head()
```

```python
# Plotting the difference in polls between Obama and Romney
fig = poll_df.plot('Start Date','Difference',figsize=(15,4),marker='o')
```

```python
#Insight: Difference observed on debate nights
#The debate dates were Oct 3rd, Oct 11, and Oct 22nd

#Finding x limits for October 2012, to zoom in and see the difference

row_index = 0
xlimit = []

#Loop to find 2012-10
#Note: this date has 7 characters thats why date[0:7]

for date in poll_df['Start Date']:
    if date[0:7] == '2012-10':
        xlimit.append(row_index)
        row_index +=1
    else:
        row_index +=1
    
print min(xlimit)
print max(xlimit)
```

```python
fig = poll_df.plot('Start Date', 'Difference', figsize= (15,4), marker = 'o', xlim = (325, 352))

#Adding markers for specific debate dates

plt.axvline(x = 325+2, linewidth = 4, color = 'purple')
plt.axvline(x = 325+10, linewidth = 4, color = 'green')
plt.axvline(x = 325 + 21, linewidth = 4, color = 'blue')
```
