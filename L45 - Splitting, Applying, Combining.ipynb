```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
#Split based on  quality
#Apply a function to it to compress it further
#Combine the diluted information together
```

```python
dframe = pd.read_csv('winequality-red.csv', sep = ';')
dframe.head()
```

```python
#Goal: Finding the highest alcohol content for each quality range

#Step 1: Sorting by descending order of alcohol

dframe.sort_values('alcohol', ascending = False, inplace = True)
dframe.head()
```

```python
#Step 2: Creating an apply function which will set ranks

def ranker(df):
    dframe['rank'] = np.arange(len(df)) + 1 
    return df
```

```python
#Step 3: Groupby to split by quality, Apply function to rank in each quality range

dframe = dframe.groupby('quality').apply(ranker)
```

```python
#Number of wines in each quality range

num_qual = dframe['quality'].value_counts()
num_qual
```

```python
#Combining: Showing the highest alcohol content in each quality range

dframe['rank' == 1].head(len(num_qual))
```
